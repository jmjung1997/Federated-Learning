```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense,Dropout,Activation
from tensorflow.python.keras import metrics
from tensorflow.python import keras
import tensorflow as tf
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
```

```python
#파일 읽기
df=pd.read_excel('./modeling_data/2weeks_summary(전체평균).xlsx')
```

```python
df
```

```python
X=df.iloc[:,2:8]
y_origin=df['Label']
y=pd.get_dummies(df['Label'])
```

```python
#Scaling
ms=MinMaxScaler()
X=ms.fit_transform(X)

# 데이터 5:5
train_x=X[:966]
test_x=X[966:]

train_y=y[:966]
test_y=y[966:]
```

```python
train_x.shape
```

```python
train_y.shape
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(6,)),
    keras.layers.Dense(128, activation=tf.nn.relu),
	keras.layers.Dense(256, activation=tf.nn.relu),
	keras.layers.Dense(128, activation=tf.nn.relu),
	keras.layers.Dense(3, activation=tf.nn.softmax),
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history=model.fit(train_x, train_y, epochs=300, batch_size=6)

predicted=model.predict(test_x)
test_loss, test_acc = model.evaluate(test_x, test_y)
```

```python
predicted.shape
```

```python
count=0
for i in range(len(test_y)):
    maxIndex=tf.argmax(predicted[i])
#     print(type(maxIndex.numpy().item()))
    index=0
    if(y_origin[i]==-1):
        index=0
    elif y_origin[i]==0:
        index=1
    elif y_origin[i]==1:
        index=2    
    if index==maxIndex.numpy().item():
        count=count+1
```

```python
print(count)
print(len(test_y))
print(count/len(test_y))
```

```python
y_vloss = history.history['val_loss']
y_loss = history.history['loss']

x_len = np.arange(len(y_loss))
plt.plot(x_len, y_vloss, marker='.', c='red', label="Validation-set Loss")
plt.plot(x_len, y_loss, marker='.', c='blue', label="Train-set Loss")

plt.legend(loc='upper right')
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```
