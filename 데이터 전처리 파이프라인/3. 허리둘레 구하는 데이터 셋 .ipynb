```python
import pandas as pd
```

```python
df=pd.read_csv("../Integrated_data/Final_dataset.csv",encoding="utf-8")
df_waist=pd.read_excel("../tutorial/data/세명초_latest.xlsx",sheet_name = 1)
```

# 허리둘레 데이터 세트

```python
df_waist=df_waist.drop([1], axis = 0)
df_waist = df_waist.rename(columns=df_waist.iloc[0])
df_waist = df_waist.drop(df_waist.index[0])
```

```python
df_waist.reset_index(inplace=True, drop=True)
```

```python
df_waist.columns=['ID','gender','before_height','before_weight','before_waist','gender','after_height','after_weight','after_waist']
```

```python
df_waist=df_waist[['ID','before_height','before_weight','before_waist','after_height','after_weight','after_waist']]
```

```python
df_waist.to_csv("../Integrated_data/waist_dataset.csv",encoding="utf-8",index=False)
```
